# Résolution par dichotomie

1-Les bibliothèques nécessaires

```python
from math import *
import matplotlib.pyplot as plt
```

2-La définiton de la fonction $f(x)=x^3-27$

```python
def f(x):
    return x**3-27
```

3-L'algorithme de résolution de $f(x_0)=0$

```python
a=0
b=5
m=(a+b)/2
epsilon=0.0001
while f(m)>epsilon:
    print("a= ",a," ; b= ",b," ; m= ",m)
    if f(a)*f(m)>0:
        a=m
        m=(a+b)/2
    else:
        b=m
        m=(a+b)/2
print("la valeur approchée de la solution est : ",m)
```

## QCM

```python
import ipywidgets as widgets
from IPython.display import display

# Définition de la question et des réponses possibles
question = "Quelle est la capitale de la France ?"
options = ["Berlin", "Madrid", "Paris", "Rome"]
correct_answer = "Paris"

# Création du widget RadioButtons pour choisir une réponse
radio_buttons = widgets.RadioButtons(
    options=options,
    description="Réponses:",
    disabled=False
)

# Fonction pour vérifier la réponse et afficher un message
def check_answer(b):
    if radio_buttons.value == correct_answer:
        print("Bonne réponse !")
    else:
        print("Mauvaise réponse. Réessaye !")

# Bouton pour soumettre la réponse
submit_button = widgets.Button(description="Valider")

# Lier le bouton à la fonction de vérification
submit_button.on_click(check_answer)

# Affichage de la question, des choix et du bouton
display(widgets.HTML(f"<h3>{question}</h3>"))
display(radio_buttons)
display(submit_button)
```
